```python
import string
import re
import numpy as np
import pandas as pd
#import sklearn.pre_processing
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
def custom_crossval(df, frac):
    '''
    Accepts a dataframe and a fraction to holdout.
     ///Returns training set and test set.
     
    '''
    
    #Take random sample of dataframe
    temp = int(len(df.index) * frac)
    print('Holdout size:', temp)
    
    rand_sample = df.sample(df.shape[0]).reset_index(drop= True)
    hold_out_set = rand_sample.loc[:temp-1]
    test_set = rand_sample.loc[temp:]
    
    
    reg = LinearRegression()
    reg.fit(test_set.loc[:,0:9], test_set.iloc[:,-1])
    print('Linear Regression R2 on test set:',
          reg.score(hold_out_set.loc[:,0:9], hold_out_set.iloc[:,-1]))

    
    return hold_out_set

holdout_frac = 0.2
size = 1000
data = pd.DataFrame(np.random.rand(size,size))
holdout_set = custom_crossval(data, holdout_frac)
holdout_set.head(10)
```
