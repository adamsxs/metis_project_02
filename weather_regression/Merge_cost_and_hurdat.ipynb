Brief summary of data sources:
* Hurricane Meteorological Data: National Oceanic and Atmospheric Administration (NOAA) HURDAT2, North Atlantic storms only. Includes time series data on tropical weather events, which has been summarized into whole-storm statistics to use as features. These storms date from the early 20th century to the present, although many are unnamed.
* Hurricane Damage Estimates: Scraped from Wikipedia pages that site post-storm reports from the NOAA's Hurricane Center. Also contains some meteorological data as well. The current import contains data from 1968 through 2017.

## Merging dataframes based on storm name and year
We will now merge the HURDAT storm characteristic data with the cost numbers scraped from Wikipedia. We have 818 entires in the cost data, and 1848 entries in the HURDAT data. Ideally, our merge should produce around 800 data points subject to other mismatches.

```python
# Converting damage_df from numeric to text for matching purposes
damage_df['year'] = damage_df.year.apply(str)

# verifying columns and dtype match. Should see 'name' and 'year'
column_match = set(damage_df.columns).intersection(set(hurdat_df.columns))
print(column_match)

print('Damage dtypes:\n', damage_df[list(column_match)].dtypes)
print('HURDAT dtypes:\n', hurdat_df[list(column_match)].dtypes)
```

```python
# Storm names repeat every 5-7 years, so merge on year as well
hur_cost = pd.merge(damage_df, hurdat_df, on=['name', 'year'])
hur_cost.info()
```

```python
#Listing columns for reference when reorganizing in cell below
print(hur_cost.columns)
hur_cost.head()
```

```python
# Making same name check as above
unq_names = Counter(hur_cost.name)
unq_names.most_common(10)
```

```python
# Move y-variable, 'damage_USD' to column 0 for easy access
# Move all non-numeric storm data to the left, grouping numeric features together
# Move all 

new_cols = hur_cost.columns.tolist()
new_cols.insert(0, new_cols.pop(6)) #move damage to front
new_cols.insert(2, new_cols.pop(10)) #move stormID forward
new_cols.insert(3, new_cols.pop(7)) #move areas_affected forward
new_cols.insert(24, new_cols.pop(10)) #move binary landfall to the back
new_cols.insert(24, new_cols.pop(11)) # move binary damage_imputed to the back


hur_cost = hur_cost.reindex(columns = new_cols)
hur_cost.info()
```

From above, we have our `y` column `damage_USD` in position 0. Numeric variate columns are in positions 5 through 24.
I'll now need to visualize the data a bit more, resolve redundant columns, and then run the regression analysis.  

Here I'm combining redundant columns and taking a quick look at feature correlation with storm damage. Nothing individually jumps out as being strongly tied, however I'm not surprised that the top values relate to storm size.

```python
# Merging the two pressure and max wind velocity columns.
# I'm going to take the lowest pressure and greatest velocity, respectively.
# This emphasizes the point of the feature, and avoids any possible Nan spillage.

#print(hur_cost.p_min.dtype, hur_cost.min_p.dtype)
#print(hur_cost.wind_v_max.dtype, hur_cost.max_sust_wind.dtype)

hur_cost['p_min'] = pd.concat([hur_cost.p_min, hur_cost.min_p], axis=1).min(axis=1)
hur_cost['wind_v_max'] = pd.concat([hur_cost.wind_v_max, hur_cost.max_sust_wind], axis=1).max(axis=1)

hur_cost.drop(labels = ['min_p', 'max_sust_wind'], axis = 1, inplace = True)

hur_cost.corr()['damage_USD'].sort_values(ascending=False)
```

Moving forward, I want to focus on a few aspects of this analysis:
* Consider removing storms that did not affect any area. I don't think that these incidents help predict damage because they violate the assumption that the storm had a chance to interact with the community. Leaving them in gives the impression of a storm that couldn't do any damage in spite of whatever strenght it has.
* There appear to be two model paths going forward:
 * Train a model only on the features where most storms have data. This would mean excluding the windspeed radii features, primarily.
 * Retain all storm records, and replace the NaN values with imputed data. Options for imputing:
  * Impute based on mean/median of that feature
  * Bin the storms if they share similar values for other features, and then impute the data based on the mean/median of that bin itself.  
* None of the features so far look to be strongly correlated with the damage values. This may change based on action on point (1) above, but if the model gets particularly bad talk with Alice about methods like piecewise-linear, where you develop different linear models for different subsections of the data.
